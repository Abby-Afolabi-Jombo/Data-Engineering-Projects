```python
#Extract the data in Json format
import requests
import pandas as pd
import json
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd"


response = requests.get(url)

data = response.json()

print(data)
```

```python
#Move Json files into Dataframes
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
# Replace NaN values with 0 so as not to lose some vital info
df.fillna(0, inplace=True)
df.info()
df
```

```python
#df["Ath_date"] = df["ath_date"].str.split("T").str[0]
#df["Ath_time"] = df["ath_date"].str.split("T").str[1]
df["Ath_time"] = df["Ath_time"].str.split(".").str[0]
```

```python
df["Last_updated_date"] = df["last_updated"].str.split("T").str[0]
df["Last_updated_time"] = df["last_updated"].str.split("T").str[1]
df["Last_updated_time"] = df["Last_updated_time"].str.split(".").str[0]
```

```python
df["Atl_date"] = df["atl_date"].str.split("T").str[0]
```

```python
df.drop(columns = ['roi','image','ath_date','atl_date','last_updated'],inplace = True)
#df.info()
```

```python
import psycopg2

# Connect to your PostgreSQL database
conn = psycopg2.connect(
    dbname="PersonalProjects_DB",
    user="postgres",
    password='--', 
    host='--', 
    port= '--'
)

# Create a cursor object to execute SQL commands
cur = conn.cursor()

# Define the SQL statement to create a table
create_table_query = '''
CREATE TABLE IF NOT EXISTS Crypto (
    id VARCHAR(50),
symbol VARCHAR(50),
name VARCHAR(50),
current_price double precision,
market_cap integer,
market_cap_rank integer,
fully_diluted_valuation double precision,
total_volume double precision,
high_h double precision,
low_h double precision,
price_change_h double precision,
price_change_percentage_h double precision,
market_cap_change_h double precision,
market_cap_change_percentage_h double precision,
circulating_supply double precision,
total_supply double precision,
max_supply double precision,
ath double precision,
ath_change_percentage double precision,
Ath_date date,
Ath_time time,
atl double precision,
atl_change_percentage double precision,
Atl_date date,
last_updated_date date,
last_updated_time time
);

'''

# Execute the SQL command to create the table
cur.execute(create_table_query)

# Commit the transaction
conn.commit()

# Close the cursor and connection
cur.close()
conn.close()

print("Table created successfully!")
```
